```python
df = pd.read_csv("cleaned_data.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
### Interpretation

- **Observation:** The distribution shows that most customers fall within the 20–35 age range, with fewer customers at higher ages.
-  The data is slightly right-skewed, indicating a concentration of younger customers.
-  This suggests the platform primarily attracts younger users, which can inform targeted marketing strategies.
```

```python
df['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
### Interpretation

-the genders are almost equally distributed with 50.3% being male and 49.7% female 
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Total Spend', data=df)
plt.title("Age vs Spending Score")
plt.show()
```

```python
### Interpretation

- **Observation:** The data points are widely scattered with no clear linear pattern.
- There is a weak or no strong correlation between age and spending score.
-  Customer spending behavior is not strongly influenced by age, suggesting other factors may be more important.
```

```python
sns.boxplot(x='Gender', y='Total Spend', data=df)
plt.title("Spending Score by Gender")
plt.show()
```

```python
### Interpretation

- The box plot shows that male customers have a higher median total spend compared to female customers. The spread of spending among males is also wider, indicating greater variability.

-  Male customers tend to spend more on average, but their spending behavior is less consistent. Female customers, on the other hand, show lower overall spending with a more concentrated distribution.

- This suggests that while male customers may contribute more to total revenue, their purchasing behavior is more varied. 
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
### Interpretation

-  The heatmap shows strong positive correlations between Total Spend, Items Purchased, and Average Rating (values close to 1). Age has a moderate negative correlation with these variables, while Discount Applied shows a strong positive correlation with Days Since Last Purchase.

- Customers who purchase more items tend to spend more and give higher ratings, indicating a consistent purchasing pattern. The negative correlation with age suggests that younger customers are more active in purchasing and spending. Additionally, higher discounts are associated with longer gaps since the last purchase.

-  The business should focus on high-engagement customers who frequently purchase and spend more, as they contribute significantly to revenue. Strategies such as targeted promotions could be used to re-engage customers who have not purchased recently, especially those responding to discounts.
```

```python
df['Gender'].value_counts().plot(kind='bar')
plt.title("Bar Chart")
plt.show()

df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Pie Chart")
plt.ylabel("")
plt.show()
```

```python
### Visualization Comparison

- **Observation:** Both charts represent the same data but in different formats.
-  The bar chart allows for easier comparison of exact values, while the pie chart emphasizes proportions.
-  Bar charts are more effective for detailed analysis, whereas pie charts are better for quick visual summaries.
```
