This dataset provides a comprehensive view of customer behaviour within an e-commerce platform. Each entryy corresponds to a unique customer with a detailed breakdown of their interactions and transactions.<br><br>

Use cases of the dataset<br>
1.<b>Customer Segmentation</b>: Analyze and categorize customers based on demographics, spending habits and satisfaction levels.<br>
2.<b>Satisfaction Analysis</b>: Investigate factorse influencing customer satisfaction and identify areas for improvement.<br>
3.<b>Promotion Strategy</b>: Assess the impact of discounts on customer spending and tailor promotional strategies accordingly.<br>
4.<b>Retention Strategies</b>: Develop targeted retention strategies by understanding the time gap since the last purchase.<br>
5.<b>City-based Insights</b>: Explore regional variations in customer behavior to optimize marketing efforts based on location-specific trends.<br>

```python
import matplotlib.pyplot as plt #Libraries for creating visuals
import seaborn as sns 

#A simple histogram showing disrtribution
plt.figure(figsize=(8,5))   
sns.histplot(df["Total Spend"], kde=True) #KDE Kernel Density Estimate shows overall shape of distribution
plt.title("Distribution of total spending of customers")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

Most customers spend about 400 to about 520. A fair amount spend more than 1400

```python
#A simple barchart

plt.bar(df["City"], df[ "Total Spend"])
plt.title("Average spending per city")
plt.xlabel("City")
plt.ylabel("Amount Spent")
plt.show()
```
