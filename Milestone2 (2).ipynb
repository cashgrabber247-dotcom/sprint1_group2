## FEATURE ENGINEERING
Feature engineering is the creation of new variables from existing data.

```python
#Simplifying the spending value of a customer to three levels for better and quicker understanding
df['Customer_Value'] = pd.cut(df['Total Spend'],
                              [0,500,1000,2000],
                              ['Low','Medium','High'])
df['Customer_Value']
```

**This feature helps distinguish between high-value and low-value customers based on spending patterns.**

```python
df['Avg_Spend_Per_Item'] = df['Total Spend'] / df['Items Purchased']
df['Avg_Spend_Per_Item']
```

**This feature helps show the average amount a customer spends on an item.**

```python
#Show how active a customer is
df['Customer_Activity'] = pd.cut(
    df['Days Since Last Purchase'],
    bins=[0, 30, 90, 365],
    labels=['Active', 'Warm', 'Inactive']
)
df['Customer_Activity']
```

**This feature shows how active a customer is and to determine the number of returning customers.**

## AGGREGATION 
Aggregation summarizes data into meaningful groups

```python
df.groupby('Membership Type')['Total Spend'].mean()
```

**Customers with higher-tier memberships such as gold, typically show higher average spending, while lower-tier memberships tend to have lower average spending.**

```python
df.groupby('City')['Total Spend'].mean()
```

**San Fransisco has highest spending indicating a strong market with high purchasing power while Houston has lowest spending, suggesting lower customer value or more price-sensitive behavior.**

```python
df.groupby('Gender')['Total Spend'].mean()
```

**Male customers have a higher average spending compared to female customers, indicating that they contribute more revenue per customer on average.
This suggests potential differences in purchasing behavior or product preferences.**

```python
df.groupby('City')['Satisfaction Level'].mean()
```

**Cities like Chicago and Miami have high satisfaction levels where as Cities such as San Fransisco and New York have low satisfaction levels as shown above where 3 represents high satisfaction and less 1 lower satisfaction.**
